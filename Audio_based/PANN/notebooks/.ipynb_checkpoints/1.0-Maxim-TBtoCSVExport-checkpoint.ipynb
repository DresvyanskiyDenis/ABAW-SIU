```python
from tensorflow.python.summary.summary_iterator import summary_iterator

def parse_summary(log_files):
    results = {}
    for ds, log_file in log_files.items():
        for summary in summary_iterator(log_file):
            if summary.step:
                if not summary.step in results:
                    results[summary.step] = {}
            
                results[summary.step]["{0}_{1}".format(ds, summary.summary.value[0].tag)] = summary.summary.value[0].simple_value
    
    return results, pd.DataFrame({epoch:datas for epoch, datas in results.items()}).T

def parse_log_files(filename, tb_log_root, is_save=False):
    logg = {
        'train': os.path.join(tb_log_root, "{0}_train".format(filename)),
        'valid': os.path.join(tb_log_root, "{0}_valid".format(filename)),
    }
    
    for idx, item in logg.items():
        logg[idx] = os.path.join(item, os.listdir(item)[0])
        
    logs, df = parse_summary(logg)
    
    if is_save:
        df.to_csv('{0}.csv'.format(filename))
        
    return df
```

```python
filename = 'mel_64_resnet18v4_grad_120_10FoldShuffle'

all_data = pd.DataFrame()
for fold in range(0, 10):    
    logg = {
        'train': os.path.join("../logs/tb/", "{0}_{1}_train".format(filename, fold)),
        'valid': os.path.join("../logs/tb/", "{0}_{1}_valid".format(filename, fold)),
    }
    
    for idx, item in logg.items():
        logg[idx] = os.path.join(item, os.listdir(item)[0])
        
    logs, df = parse_summary(logg)
    
    all_data['train_recall_{0}'.format(fold)] = df['train_recall']
    all_data['valid_recall_{0}'.format(fold)] = df['valid_recall']
    
all_data.to_csv('{0}.csv'.format(filename))
```
